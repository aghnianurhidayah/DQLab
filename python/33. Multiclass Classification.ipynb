## Proses Multiclass Classification


<b>1. Siapkan dataset, sekumpulan data (X) beserta label untuk setiap data (y).</b>



- Pada gambar di atas, setiap data yang menyusun X direpresentasikan oleh baris berwarna merah dan label yang dimilikinya direpresentasikan dengan kolom warna biru pada baris tersebut.
- Karena sebuah data terdiri dari sekumpulan properti (list of properties) dan X terdiri dari sekumpulan data (list of data), maka dapat disimpulkan bahwa X merupakan kumpulan dari sekumpulan properti (list of list of properties).Hal ini membuat X dapat direpresentasikan sebagai matriks atau tipe data list dua dimensi.
- Di sisi lain, dikarenakan y terdiri dari sekumpulan data dimana tiap data hanya terdiri satu nilai label saja maka dapat disimpulkan bahwa y merupakan kumpulan dari nilai-nilai label (list of labels).
- Sehingga, masukan pertama berupa matriks dan masukan kedua berupa list.
- Kemudian, saat model telah selesai dilatih, dapat dispesifikasikan dari sekumpulan data yang ingin diprediksi labelnya dalam bentuk matriks sebagai masukan untuk model dan model akan mengembalikan sekumpulan label untuk setiap data dalam bentuk list.
- Kolom target, y, pada tugas klasifikasi multi-kelas kemungkinan nilai yang dapat dimiliki oleh kolom lebih dari dua. Pada contoh terdapat 4 kemungkinan kelas yang dapat dimiliki oleh suatu data yaitu 0, 1, 2, dan 3.


<b>2. Bagi dataset ke dalam dua bagian (dataset training dan dataset testing).</b>

Dataset training akan digunakan untuk melatih model. Dataset testing akan digunakan untuk mengevaluasi performa dari model untuk data yang belum pernah dijumpai.

<b>3. Inisialisasi model klasifikasi yang ingin digunakan beserta nilai parameter milik model.</b>

Catatan:
- jika belum memiliki gambaran terkait dengan nilai parameter yang dimiliki oleh model, maka dapat menggunakan default parameter yang ditawarkan.
- jumlah dari parameter yang dimiliki oleh suatu model klasifikasi beragam (ada model yang bahkan tidak memiliki parameter dan ada pula model yang memiliki banyak parameter).
- mencoba dua buah model klasifikasi yaitu Multinomial Naive Bayes dan k-Nearest Neighbour serta membahas parameter yang dapat kita gunakan untuk mempengaruhi performa model.

<b>4. Latih model dengan dataset training (hasil dari tahap 2).</b>

<b>5. Ukur performa model berdasarkan dataset training dan data testing (hasil dari tahap 2) untuk melakukan evaluasi terhadap model hasil latih.</b>

Proses perhitungan dapat dilakukan melalui langkah-langkah berikut.

1. Lakukan prediksi pada sekumpulan data X (baik milik dataset training ataupun dataset testing) untuk menghasilkan list hasil prediksi
2. Bandingkan hasil prediksi dengan label aktual yang dimiliki oleh setiap data.
3. Hitung jumlah prediksi yang sama dengan label aktualnya
4. Bagi hasil perhitungan dengan total data X untuk menghasilkan akurasi



Strategi K-Fold bekerja dengan membagi data ke sejumlah K kelompok. Kemudian, berdasarkan proses pembagian data, akan dilakukan proses percobaan pelatihan dan evaluasi model dengan menggunakan satu kelompok data sebagai data testing dan kelompok data lainnya sebagai data training. Proses tersebut akan dilakukan terus menerus hingga seluruh kelompok data kebagian menjadi data testing. 
 
Strategi K Fold dapat memanfaatkan seluruh data yang dimiliki  sebagai data uji dan mengetahui nilai performa model yang lebih representatif (tidak bias terhadap suatu konfigurasi pemilihan data testing dan training).

Metode Multinomial Naive Bayes bekerja dengan menggunakan persamaan berikut ini.

#### Kelebihan dan Kekurangan Algoritma Multinomial Naive Bayes
<b>Kelebihan Algoritma Multinomial Naive Bayes</b>

1. Algoritma Multinomial Naive Bayes baik digunakan saat seluruh/ sebagian besar atribut input yang ingin dipelajari bersifat mutually independen (nilai dari setiap atribut tidak dipengaruhi oleh nilai atribut lainnya). Sebagai contoh: atribut jenis kelamin dan usia (jenis kelamin seseorang tidak mempengaruhi usia yang ia miliki)
2. Algoritma Multinomial Naive Bayes tidak membutuhkan data training dalam jumlah yang besar dengan asumsi syarat pertama terpenuhi.
3. Saat data masukan yang ingin dipelajari tersusun atas atau sebagian besarnya berupa data kategori algoritma Multinomial Naive Bayes dapat bekerja dengan sangat baik jika dibandingkan saat data yang ingin dipelajari berupa data numerik.
4. Algoritma ini juga memiliki tingkat kompleksitas komputasi yang lebih rendah jika dibandingkan dengan algoritma pembelajaran mesin lainnya sehingga proses pelatihan dari algoritma cenderung cepat jika dibandingkan dengan algoritma lainnya yang lebih kompleks seperti metode ensemble (Random Forest, Gradient Boosting Classifier, dll) ataupun algoritma Multilayer Perceptron. Kemudian, algoritma ini juga tidak memiliki parameter yang harus di-adjust untuk menghasilkan performa yang baik sehingga proses pengembangan dan evaluasi model dapat berlangsung dengan lebih cepat.


<b>Kekurangan Algoritma Multinomial Naive Bayes</b>

1. Saat sebagian besar atau bahkan seluruh atribut input yang ingin dipelajari tidak bersifat mutually independent, nilai kemunculan setiap atribut sangat berpengaruh terhadap atribut lainnya maka algoritma akan memiliki performa yang sangat buruk. Sebagai contoh: total pendapatan dan persentase pajak penghasilan (semakin besar total pendapatan maka akan semakin besar pula persentase pajak penghasilan)
2. Saat jumlah data yang ingin dipelajari berukuran sangat besar maka akan semakin besar pula kemungkinan suatu atribut yang ingin dipelajari bergantung pada atribut lainnya sehingga algoritma tidak dapat bekerja secara maksimal.
3. Algoritma cenderung memiliki performa yang buruk untuk data numerik dibandingkan dengan data kategori.
4. Algoritma cenderung memiliki performa yang tidak terlalu baik pada dataset dalam jumlah yang besar jika dibandingkan dengan algoritma yang lebih kompleks seperti metode ensemble (Random Forest, Gradient Boosting Classifier, dll) ataupun algoritma Multilayer Perceptron.

### Metode k-Nearest Neighbors

Metode k-Nearest Neighbors (k-NN) merupakan metode klasifikasi yang paling sederhana. #### Kelebihan dan Kekurangan Algoritma k-Nearest Neighbors

<b>Kelebihan dari algoritma k-NN:</b>

1. Dikarenakan prinsip kerja dari algoritma k-NN yang sederhana, algoritma mudah untuk dipahami dan mudah untuk diimplementasikan (hanya parameter jumlah tetangga, n_neighbors, dan fungsi jarak yang harus di-adjust dalam proses parameterisasi model - fine-tuning model)
2. Saat dimensi data yang ingin dipelajari tidak terlalu besar dan jumlah data relatif sedikit, algoritma kNN mampu mempelajari data dalam waktu yang cepat dan performa yang baik.
3. Algoritma cenderung memiliki performa yang baik ketika data input memiliki tipe atribut yang homogen (seluruh tipe atribut sama, e.g. seluruhnya berupa fitur binary atau seluruhnya berupa atribut numerik).


<b>Kekurangan dari algoritma k-NN:</b>

1. Dikarenakan algoritma k-NN bekerja dengan menggunakan suatu fungsi jarak, algoritma ini pun turut terkena fenomena kutukan dimensi (curse of dimensionality). Hal ini mengartikan bahwa saat atribut input yang ingin dipelajari berjumlah besar dan jumlah data yang tersedia sedikit maka persebaran data menjadi sangat jarang yang mengakibatkan  algoritma k-NN menjadi tidak efektif (menghasilkan performa yang buruk).
2. Algoritma cenderung memiliki performa yang buruk ketika data input memiliki tipe atribut yang bersifat heterogen (tipe atribut yang muncul beragam, e.g. tipe data berupa data kategori biner, kategori ordinal, kategori ordinal dan numerik)
3. Algoritma tidak cocok digunakan untuk mempelajari data latih yang berukuran besar dikarenakan saat proses inferensi dilakukan algoritma bekerja berdasarkan proses pengukuran jarak pada setiap data dalam data latih, sehingga semakin besar ukuran data latih, akan semakin lama pula proses inferensi berlangsung (data yang ingin diprediksi harus dibandingkan dengan setiap data dalam data latih).
