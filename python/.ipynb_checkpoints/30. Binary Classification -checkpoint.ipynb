## Proses Binary Classification




<b>1. Siapkan dataset, sekumpulan data (X) beserta label untuk setiap data (y).</b>
Perhatikan bahwa setiap data memiliki sekumpulan properti ilustrasi dari data, label serta properti yang disiapkan dapat dilihat seperti gambar di bawah ini.



- Pada gambar di atas, setiap data yang menyusun X direpresentasikan oleh baris berwarna merah dan label yang dimilikinya direpresentasikan dengan kolom warna biru pada baris tersebut. Sesuai dengan ilustrasi pada gambar, karena sebuah data terdiri dari sekumpulan properti (list of properties) dan X terdiri dari sekumpulan data (list of data), maka dapat disimpulkan bahwa X merupakan kumpulan dari sekumpulan properti (list of list of properties).
- Hal ini membuat X dapat direpresentasikan sebagai matriks atau menggunakan tipe data list dua dimensi, numpy array 2D, dan pandas dataframe. Di sisi lain, dikarenakan label hanya direpresentasikan dengan satu kolom dengan nilai nol atau satu di dalamnya maka dapat disimpulkan bahwa y merupakan kumpulan dari angka satu dan nol (list of zero and one). Untuk itu, y yang sebagai vektor dapat direpresentasikan ke dalam tipe data list, numpy array 1D, atau pandas series.
- Saat model klasifikasi biner dilatih, model akan menerima dua buah masukan, di mana masukan pertama berupa matriks dan masukan kedua berupa list atau vektor.
- Saat model telah selesai dilatih, baru dapat menspesifikasikan sekumpulan data yang ingin diprediksi labelnya dalam bentuk matriks sebagai masukan untuk model dan model akan mengembalikan sekumpulan label untuk setiap data dalam bentuk list atau vektor.

<b>2. Bagi dataset ke dalam dua bagian (dataset training dan dataset testing).</b>

Dataset training akan digunakan untuk melatih model. Sedangkan dataset testing akan digunakan untuk mengevaluasi performa dari model untuk data yang belum pernah dijumpai.

<b>3. Inisialisasi model klasifikasi yang ingin digunakan beserta nilai parameter milik model.</b>

Catatan:
- jika belum memiliki gambaran terkait dengan nilai parameter yang dimiliki oleh model maka dapat menggunakan default parameter yang ditawarkan;
- jumlah dari parameter yang dimiliki oleh suatu model klasifikasi beragam (ada model yang bahkan tidak memiliki parameter dan ada pula model yang memiliki banyak parameter);
- Modul ini akan mencoba dua buah model klasifikasi yaitu Decision Tree dan Random Forest serta membahas parameter yang dapat digunakan untuk mempengaruhi performa model.

<b>4. Latih model dengan dataset training (hasil dari tahap 2).</b>

<b>5. Untuk mengukur performa hasil prediksi model, dapat dihitung akurasi dari model hasil pelatihan.</b>

Proses perhitungan dapat dilakukan melalui langkah-langkah berikut.
- Lakukan prediksi pada sekumpulan data X (baik milik dataset training ataupun dataset testing) untuk menghasilkan list hasil prediksi
- Bandingkan hasil prediksi dengan label aktual yang dimiliki oleh setiap data.
- Hitung jumlah prediksi yang sama dengan label aktualnya.
- Bagi hasil perhitungan dengan total data X untuk menghasilkan akurasi. 


### Algoritma Decision Tree

Algoritma Decision Tree bekerja dengan menggunakan prinsip divide and conquer untuk memisahkan suatu dataset menjadi sekelompok sub-dataset berdasarkan atribut yang dimilikinya secara rekursif.

Model decision tree bekerja dengan memisahkan dataset ke sub-dataset yang lebih seragam (homogenous) secara rekursif. Proses pemisahan dataset dapat dilakukan dengan memilih suatu titik potong (cutting point) yang membuat dataset menjadi lebih seragam.



Gambar di atas merepresentasikan sekumpulan data yang memiliki dua buah atribut (Atribut 1 dan Atribut 2) dengan dua macam kelas (label) yang dapat dimiliki oleh setiap data tersebut (Biru dan Merah).

### Algoritma Random Forest

Proses pelatihan metode Random Forest bisa dilakukan melalui tahap-tahap berikut

1. Menerima n sebagai jumlah dari Decision Tree yang akan dibentuk beserta parameter/ kriteria dari setiap Decision Tree yang akan dibentuk (e.g. max_depth dan min_samples_split) 
2. Menerima dataset X beserta label y yang ingin dipelajari
3. Inisialisasi variabel kosong RF untuk menampung seluruh Decision Tree yang akan terbentuk
4. Melakukan proses subsampling (berdasarkan fitur dan data) sebanyak n kali dan untuk setiap subsampel data lakukan:
    1. Proses pelatihan model Decision Tree menggunakan subsampel
    2. Tambahkan model yang telah dilatih ke variabel RF
    
Selanjutnya untuk memprediksi data baru, metode Random Forest bekerja dengan melakukan proses prediksi berdasarkan setiap Decision Tree yang berada dalam variabel RF dan melakukan proses majority voting untuk menentukan label dari data tersebut
